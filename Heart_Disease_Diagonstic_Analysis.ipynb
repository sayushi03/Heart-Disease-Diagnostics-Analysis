```python
data['Heart_disease_sex']=''
```

```python
def Heart_disease_sex(sex,target):
  if sex==0 & target==1:
    return 1
  elif sex==1 & target==1:
    return 2
  elif sex==0 & target==0:
    return 0
  else:
    return 0
```

```python
data['Heart_disease_sex']=data.apply(lambda row: Heart_disease_sex(row['sex'],row['target']),axis=1)
```

```python
def rest_ecg(restecg,target):
  if restecg==0 & target==1:
    return 'Normal Type'
  elif restecg==1 & target==1:
    return 'ST wave Abnormality'
  elif restecg==2 & target==1:
    return 'showing probable or definite left ventrical hypertrophy'
  else:
    return 'Absent'
```

```python
data['rest_ecg']=''
```

```python
data['rest_ecg']=data.apply(lambda row: rest_ecg(row['restecg'],row['target']),axis=1)
```
